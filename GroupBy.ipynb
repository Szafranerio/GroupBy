```python
import pandas as pd
```

```python
data = pd.read_csv('fortune1000.csv',index_col='Rank')
data.head()
```

```python
sectors = data.groupby('Sector')
sectors
```

```python
len(sectors)
```

```python
#List of scetors with total amount of each sector
sectors.size()
```

get_group method

```python
#Gives overview of whole group!!
sectors.get_group('Energy')
```

```python
import seaborn as sns
```

```python
#There is a possibility to make whole analysis based on grouped values.
sns.pairplot(sectors.get_group('Energy'))
```

```python
#Grouping allows to perform statistics faster
sectors['Revenue'].sum()
```

```python
sectors['Employees'].describe()
```

The agg (aggregation) method 

```python
#Results is a dataframe
sectors.agg({'Revenue':'sum', 'Profits':'max'})
```

Iterating thorugh groups

```python
#Two rows of value with the higest profits, in the bracket numeric value points the amount of the rows to show.
data.nlargest(2, 'Profits')
```

```python
def top_two_companies_by_employee_count(sector):
    return sector.nlargest(2, 'Employees')

sectors.apply(top_two_companies_by_employee_count)
```
